##### Genre based Analysis

```python
df_genre = pd.DataFrame(columns = ['genre', 'budget', 'gross', 'year', 'other_genres'])

def genreRemap(row):
    global df_genre
    d={}
    genres = np.array(row['genres'].split('|'))
    n = genres.size
    d['budget'] = [row['budget']]*n
    d['gross'] = [row['gross']]*n
    d['year'] = [row['title_year']]*n
    d['genre'], d['other_genres'] = [], []
    
    for g in genres:
        d['genre'].append(g)
        d['other_genres'].append(genres[genres != g])
    
    df_genre = df_genre.append(pd.DataFrame(d), ignore_index=True)
    
df_clean_genres = df1[['budget', 'genres', 'gross', 'title_year']].dropna()
df_clean_genres.head(3)

df_clean_genres.apply(genreRemap, axis=1)
df_genre['year'] = df_genre['year'].astype(np.int16)
df_genre = df_genre[['genre', 'budget', 'gross', 'year', 'other_genres']]

df_genre.head(5)
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.decomposition import PCA # Principal Component Analysis module
from sklearn.cluster import KMeans # KMeans clustering 
import matplotlib.pyplot as plt # Python defacto plotting library
import seaborn as sns # More snazzy plotting library
%matplotlib inline
```

```python
str_list = [] # empty list to contain columns with strings (words)
for colname, colvalue in df1.iteritems():
    if type(colvalue[1]) == str:
         str_list.append(colname)
# Get to the numeric columns by inversion            
num_list = df1.columns.difference(str_list)
```

```python
movie_num = df1[num_list]
movie_num.head()
```

```python
movie_num = movie_num.fillna(value=0, axis=1)
```

```python
# calculating Eigenvectors and eigenvalues of cov matrix
mean_vec = np.mean(X_std,  axis=0)
cov_mat = np.cov(X_std.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
```

```python
# create a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:,i]) for i in range(len(eigen_vals))]

# sorting high to low
eig_pairs.sort(key = lambda x : x[0], reverse=True)

#calculating explained variance from eigen values
tot = sum(eigen_vals)
variance_exp = [(i/tot)*100 for i in sorted(eigen_vals, reverse=True)]
cum_var_exp = np.cumsum(variance_exp)

# plotting explained variance
plt.figure(figsize=(8,5))
plt.bar(range(16), variance_exp, alpha=0.33, align='center', label='Explained variance', color='g')
plt.step(range(16), cum_var_exp, where='mid', label='cumulative variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('principle components')
plt.legend(loc='upper left')
plt.show()
```

##### Principal component analysis

```python
pca = PCA(n_components=3)
x_3 = pca.fit_transform(X_std)
y = df1['imdb_score']

from mpl_toolkits.mplot3d import Axes3D
plt.clf
fig = plt.figure(1, figsize=(8,8))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(x_3[:,0], x_3[:,1], x_3[:,2], c=y, cmap=plt.cm.Paired)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvertor")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
